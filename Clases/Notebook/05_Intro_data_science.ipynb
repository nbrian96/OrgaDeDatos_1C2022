# **scikit-learn**

```
import numpy as np
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

amount = 100
#generamos aleatoriamente 100 números entre 0 y 1, este vector tendrá los números en X
x_data = np.random.rand(amount).astype(np.float32)
#definimos un nuevo vector que son los números en Y (dados por la ecuación y=x*3+2)
y_data = x_data * 3 + 2
#Esto agrega ruido a los números en X
y_data = np.vectorize(lambda y: y + np.random.normal(loc=0.0, scale=0.1))(y_data)
#si quieren graficar los puntos, ejecuten las siguientes dos líneas:
plt.scatter(x_data,y_data)
plt.show()
```

**Implementación basica, Python puro**

```
x = 0
y = 0
xy =0
xx = 0

for i in range(0, amount):
  x += x_data[i];
  y += y_data[i];
  xy += x_data[i]*y_data[i];
  xx += x_data[i]*x_data[i];


calculated_b = ((amount * xy) - (x * y)) / ((amount * xx) - (x * x));

calculated_a = (y - (calculated_b * x)) / amount;

print(calculated_a)
print(calculated_b)

if (calculated_b != 0): 
   
  calculated_y = x_data*calculated_b+calculated_a  
  plt.plot(x_data, y_data, 'ro', color="blue")
  plt.plot(x_data, calculated_y, '-r', label='linear regression', color="red")  
  plt.show()

else:
     print('No linear solution');

```

**`Usando SkLearn`**

Desde el punto de vista de la implementación, esto es simplemente Mínimos cuadrados ordinarios (scipy.linalg.lstsq) o Mínimos cuadrados no negativos (scipy.optimize.nnls) envueltos como un objeto predictor.

```
from sklearn.linear_model import LinearRegression
import numpy as np
# # Regresor lineal
# modelo = LinearRegression()
# # Se entrena el modelo con los datos
# modelo.fit(x_data, y_data)

x_data_np = np.array(x_data).reshape(-1, 1)
y_data_np = np.array(y_data)

reg = LinearRegression().fit(x_data_np, y_data_np)
print(reg.score(x_data_np, y_data_np))

print(reg.coef_)
# Una vez ha sido entrenado, se puede calcular el resultado
# para una nueva entrada


Y_prediccion = reg.predict(x_data_np)

plt.plot(x_data, y_data, 'ro', color="blue")
plt.plot(x_data_np, Y_prediccion, '-r', label='linear regression', color="red")  
plt.show()
```

**Regresión Logistica**

```
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
list(iris.keys())
```

```
X=iris["data"][:,3:] #anchura del petalo, (para cada fila tomar el elemento de la columna 3)
y = (iris["target"] == 2).astype(int) #1 si es Iris virginica, 0 si no
```

```
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X,y)
```

Ahora veamos como predice un nuevo conjunto de datos.
Flores con anchuras de petalos entre 0cm y 3 cm

```
import matplotlib.pyplot as plt
X_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:,1], "g-", label = "Iris virginica")
plt.plot(X_new, y_proba[:,0], "r--", label = "No es Iris virginica")
```

**K-Means**

```
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
list(iris.keys())
```

```
from sklearn.cluster import KMeans
k = 3
kmeans = KMeans(n_clusters=k)
X = iris.data
y = iris.target
y_pred = kmeans.fit(X)
```

**Representación 2D de los datos, según la clasificación ya conocida**

```
# The indices of the features that we are plotting
x_index = 0
y_index = 1

# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])

plt.tight_layout()
plt.show()
```

**Representación 2D de los datos, según los clusters de K-Means**

```
# The indices of the features that we are plotting
x_index = 0
y_index = 1

predicted = kmeans.predict(X)
clusterNames = ["cluster 0", "cluster 1", "cluster 2"]

# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: clusterNames[int(i)])

plt.figure(figsize=(5, 4))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=predicted)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])

plt.tight_layout()
plt.show()
```
